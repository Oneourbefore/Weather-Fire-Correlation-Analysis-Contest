## 강원도에서 중복행 제거

```python
# 중복된 행 찾기 (모든 중복 행 포함)
duplicates = df_gang[df_gang.duplicated(subset=['tm', 'district_1', 'district_2', 'stn'], keep=False)]

# -99의 개수를 세는 함수
def count_neg99(row):
    return (row == -99).sum()

# 중복된 값들끼리 그룹화하여 -99가 많은 행 제거
def remove_row_with_most_neg99(group):
    if len(group) > 1:
        neg99_counts = group.apply(count_neg99, axis=1)
        # -99가 가장 많은 행의 인덱스 찾기
        max_neg99_index = neg99_counts.idxmax()
        # -99가 가장 많은 행을 제거
        group = group.drop(index=max_neg99_index)
    return group

# 중복되지 않은 행과 중복된 행 중 -99가 더 많은 행 제거 후 결합
df_no_duplicates = df_gang[~df_gang.duplicated(subset=['tm', 'district_1', 'district_2', 'stn'], keep=False)]
grouped_duplicates = duplicates.groupby(['tm', 'district_1', 'district_2', 'stn']).apply(remove_row_with_most_neg99).reset_index(drop=True)
df_gang = pd.concat([df_no_duplicates, grouped_duplicates]).reset_index(drop=True)
df_gang
```

## stn이 121인 것(영월)을 제외하고 이상치를 평균 값으로 처리

```python
# stn이 121인 행 제외
df_filtered = df_gang[df_gang['stn'] != 121]

# 제외할 열 목록
exclude_columns = ['tm', 'district_1', 'district_2']

# -99를 각 열의 평균값으로 대체하는 함수
def replace_neg99_with_mean(group):
    for column in group.columns:
        if column not in exclude_columns and column != 'stn':
            mean_value = group[group[column] != -99][column].mean()
            group[column] = group[column].replace(-99, mean_value)
    return group

# 각 stn별로 그룹화하여 -99 값을 대체
df_replaced = df_filtered.groupby('stn').apply(replace_neg99_with_mean).reset_index(drop=True)
df_replaced
```

```python
# 이상치 확인

check = df_replaced.apply(lambda row: (row == -99).any(), axis=1).groupby(df_replaced['stn']).sum()
check
```

```python
# stn이 121인 행 필터링
df_stn_121 = df_gang[df_gang['stn'] == 121]

# 각 열에서 -99의 개수 확인
column_neg99_counts = (df_stn_121 == -99).sum()

# 행별로 -99의 개수 확인
row_neg99_counts = df_stn_121.apply(lambda row: (row == -99).sum(), axis=1)

print("각 열에서 -99의 개수:")
print(column_neg99_counts)

print("\n행별로 -99의 개수:")
print(row_neg99_counts)
```
