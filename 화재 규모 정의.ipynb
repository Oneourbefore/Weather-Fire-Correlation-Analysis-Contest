```python
import matplotlib.pyplot as plt # 시각화를 위한 라이브러리
import seaborn as sns # 시각화를 위한 라이브러리
import numpy as np # 다차원 배열을 위한 라이브러리
import warnings
warnings.filterwarnings(action="ignore")
import pandas as pd
```

```python
df_fire = pd.read_csv("/content/drive/MyDrive/날씨 빅데이터 공모전/data/gangwon_fire.csv")
df_fire = df_fire.drop("Unnamed: 0", axis = 1)
df_fire
```

```python
df_fire = df_fire.drop(['dead', 'injury'], axis=1)
df_fire
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
filtered_df = df_fire[df_fire['property_damage'] != 0]

# 표준화
scaler = StandardScaler()
data = filtered_df[['property_damage']]
scaled_data = scaler.fit_transform(data)

# 클러스터 개수 범위
k_range = range(1, 11)
sse = []

# 각 클러스터 개수에 대해 K-means
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=20, n_init=20)
    kmeans.fit(scaled_data)
    sse.append(kmeans.inertia_)

# Elbow 메서드를 이용하여 최적 클러스터 개수 결정
plt.figure(figsize=(8, 6))
plt.plot(k_range, sse, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Sum of Squared Errors (SSE)')
plt.title('Elbow Method for Optimal K')
plt.grid(True)
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=20, n_init=20)


kmeans.fit(scaled_data)


cluster_labels = kmeans.labels_


df_fire['Level_init_property'] = 0


filtered_df['Level_init_property'] = cluster_labels + 1  # 클러스터 레이블이 0부터 시작하므로 1을 더해줌

# 원래 데이터에 클러스터 레이블 병합
df_fire.update(filtered_df[['Level_init_property']])
df_fire
```

```python
df_fire['Level_init_property'].value_counts()
```

```python
non_zero_df = df_fire[df_fire['property_damage'] > 0]

# property_damage 열을 기준으로 정렬
df_sorted = non_zero_df.sort_values(by='property_damage')

# 각 구간의 경계값 설정 (백분위수 계산)
percentile_25 = df_sorted['property_damage'].quantile(0.25)
percentile_50 = df_sorted['property_damage'].quantile(0.50)
percentile_75 = df_sorted['property_damage'].quantile(0.75)

# level 열 계산
def calculate_level(value):
    if value <= percentile_25:
        return 1
    elif value <= percentile_50:
        return 2
    elif value <= percentile_75:
        return 3
    elif value < 100000:
        return 4
    else:
        return 5

# level 열 추가 (0인 값은 0으로 설정)
df_fire['level_property'] = df_fire['property_damage'].apply(lambda x: 0 if x == 0 else calculate_level(x))
df_fire
```

```python
property_damage_stats = df_fire['property_damage'].describe()

# 기술 통계 요약을 일반적인 숫자 형태로 변환하여 출력
property_damage_stats_str = property_damage_stats.apply(lambda x: f"{x:.0f}")

print(property_damage_stats_str)
```

```python
filtered_df_2 = df_fire[df_fire['casualties'] != 0]


scaler = StandardScaler()
data = filtered_df_2[['casualties']]
scaled_data = scaler.fit_transform(data)


k_range = range(1, 11)
sse = []


for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=20, n_init=20)
    kmeans.fit(scaled_data)
    sse.append(kmeans.inertia_)

# Elbow 메서드를 이용하여 최적 클러스터 개수 결정
plt.figure(figsize=(8, 6))
plt.plot(k_range, sse, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Sum of Squared Errors (SSE)')
plt.title('Elbow Method for Optimal K')
plt.grid(True)
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=20, n_init=20)


kmeans.fit(scaled_data)


cluster_labels = kmeans.labels_


df_fire['Level_init_casualties'] = 0


filtered_df_2['Level_init_casualties'] = cluster_labels + 1  # 클러스터 레이블이 0부터 시작하므로 1을 더해줌

# 원래 데이터에 클러스터 레이블 병합
df_fire.update(filtered_df_2[['Level_init_casualties']])
df_fire
```
